# Logistic regression

##### Logistic regression is a statistical method used for binary classification, predicting the probability that a given input belongs to one of two categories

```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.linear_model import LogisticRegression
```

```python
tit=pd.read_csv(r"C:\Users\diyan\Downloads\Telegram Desktop\Titanicc.csv")
tit
```

```python
tit.isna().sum()
```

```python
mean=tit.Age.mean()
tit['Age']=tit['Age'].fillna(mean)
```

```python
tit=tit.drop(['Cabin','PassengerId','Name','Ticket',],axis=1)
tit
```

```python
mode=tit['Embarked'].mode()[0]
tit['Embarked']=tit['Embarked'].fillna(mode)
```

```python
tit.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
tit['Sex']=tit['Sex'].astype(str)
tit['Sex']=LabelEncoder().fit_transform(tit['Sex'])
```

```python
tit['Embarked']=LabelEncoder().fit_transform(tit['Embarked'])
```

```python
tit
```

```python
y=tit[['Survived']]
y
```

```python
x=tit.drop(['Survived'],axis=1)
x
```

```python
model=LogisticRegression()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
test_pred=model.predict(x_test)
test_pred
```

```python
r2_score(test_pred,y_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(test_pred,y_test)
```

```python
confusion_matrix(test_pred,y_test)
```
