```python
mean=hire['test_score(out of 10)'].mean()
hire['test_score(out of 10)']=hire['test_score(out of 10)'].fillna(mean)
```

```python
hire
```

```python
y=hire[['salary($)']]
y
```

```python
x=hire.drop(['salary($)'],axis=1)
x
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
correlation=hire.corr()
correlation
```
